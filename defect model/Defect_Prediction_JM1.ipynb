## Cek Outliers

```python
outliers_sum = 0
outlier_mask = pd.Series(False, index=df.index)

# Iterasi untuk setiap kolom kecuali kolom terakhir
for col in df.columns[:-1]:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Identifikasi outliers
    outliers = ((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))).sum()

    # Print jumlah outliers pada setiap kolom
    print(f"Number of outliers in '{col}' column: {outliers}")
    outliers_sum += outliers

    # Tandai baris yang mengandung outliers
    outlier_mask |= (df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))

# Print jumlah total outliers
print(f"\nTotal number of outliers (across all rows and columns): {outliers_sum}")

# Jumlah baris yang mengandung setidaknya satu outlier
rows_with_outliers = outlier_mask.sum()
print(f"Number of rows with at least one outlier: {rows_with_outliers}")
```

# Feature Selection

```python
class SSO:
    def __init__(self, obj_func, n_features, pop_size=20, epochs=50):
        self.obj_func = obj_func
        self.n_features = n_features
        self.pop_size = pop_size
        self.epochs = epochs

        # Inisialisasi populasi
        self.population = np.random.randint(2, size=(pop_size, n_features))
        self.fitness = np.array([self.obj_func(ind) for ind in self.population])
        self.g_best = self.population[np.argmin(self.fitness)].copy()
        self.g_best_fitness = np.min(self.fitness)
        self.fitness_history = []

    def evolve(self):
        """
        Evolusi Salp Swarm Optimization.
        """
        self.fitness_history = []
        for epoch in range(self.epochs):
            c1 = 2 * np.exp(-((4 * epoch / self.epochs) ** 2))  # Eq. (3.2)
            new_population = []
            epoch_fitness = []
            for i in range(self.pop_size):
                if i == 0:  # Pemimpin salp
                    c2 = np.random.rand(self.n_features)
                    c3 = np.random.rand(self.n_features)
                    leader_pos1 = self.g_best + c1 * (np.random.rand(self.n_features) - 0.5)
                    leader_pos2 = self.g_best - c1 * (np.random.rand(self.n_features) - 0.5)
                    new_pos = np.where(c3 < 0.5, leader_pos1, leader_pos2)
                else:  # Pengikut salp
                    new_pos = (self.population[i] + self.population[i - 1]) / 2

                # Konversi ke biner menggunakan sigmoid
                new_pos = (1 / (1 + np.exp(-new_pos)) > np.random.rand(self.n_features)).astype(int)

                # Evaluasi solusi baru
                fitness = self.obj_func(new_pos)
                epoch_fitness.append(fitness)
                if fitness < self.g_best_fitness:  # Update global best
                    self.g_best = new_pos.copy()
                    self.g_best_fitness = fitness

                new_population.append(new_pos)

            self.population = np.array(new_population)
            self.fitness = np.array(epoch_fitness)

            # Simpan fitness terbaik dari epoch ini ke history
            best_fitness_epoch = np.min(epoch_fitness)
            self.fitness_history.append(best_fitness_epoch)

            print(f"Epoch {epoch + 1}/{self.epochs}, Best Fitness This Epoch: {best_fitness_epoch:.4f}, Overall Best Fitness: {self.g_best_fitness:.4f}")
        return self.g_best, self.g_best_fitness, self.fitness_history
```

```python
def objective_function(solution):
    selected_features = [i for i, bit in enumerate(solution) if bit == 1]

    # Penalti jika tidak ada fitur dipilih
    if len(selected_features) == 0:
        return 1e6  # Penalti besar

    # Subset fitur
    X_train_selected = X_train.iloc[:, selected_features]
    X_test_selected = X_test.iloc[:, selected_features]

    # Random Forest untuk evaluasi
    model = RandomForestClassifier(random_state=42)
    model.fit(X_train_selected, y_train)
    y_pred = model.predict(X_test_selected)

    # Akurasi
    accuracy = accuracy_score(y_test, y_pred)

    # Penalti jumlah fitur
    feature_ratio = len(selected_features) / X_train.shape[1]

    # Gabungkan akurasi dan penalti
    alpha = 0.99  # Bobot akurasi
    gamma = 1 - alpha  # Bobot penalti
    fitness = -(alpha * accuracy + gamma * (1 - feature_ratio))  # Negasi untuk minimisasi
    return fitness
```

```python
# Inisialisasi SSO
sso = SSO(objective_function, n_features=X_train.shape[1], pop_size=40, epochs=50)

# Optimasi
best_solution, best_fitness, fitness_record = sso.evolve()

# Hasil
selected_features = [i for i, bit in enumerate(best_solution) if bit == 1]
print(f"Selected Features: {selected_features}")
print(f"Best Fitness: {best_fitness}")
```

```python
fitness_record = [abs(x) for x in fitness_record]
fitness_record
```

# Modelling and Evaluation With Selected Feature

## Modelling

```python
model = RandomForestClassifier(n_estimators=200, max_depth=20, random_state=42,class_weight={0:3,1:5})
scores = cross_val_score(model, X_train_selected, y_train, cv=5, scoring='accuracy')
print(f"Cross-validation scores: {scores}")
print(f"Mean accuracy: {np.mean(scores)}")
```

```python
model.fit(X_train_selected, y_train)
y_pred = model.predict(X_test_selected)
```

## Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Evaluate the model (With bSSA)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)


sns.heatmap(cm, annot=True, fmt='g')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Assuming you have y_test and y_pred from the previous code
y_pred_proba = model.predict_proba(X_test_selected)[:, 1]  # Probability of positive class

fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

print(f"AUC = {roc_auc}")
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
